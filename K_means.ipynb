```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv("Mall_Customers.csv")
print(df.head())
```

```python
x = df.iloc[:,[3,4]].values
```

```python
wcss_list = []

for i in range(1,11):
  k_means = KMeans(n_clusters = i,init = 'k-means++',random_state  = 42 )
  k_means.fit(x)
  wcss_list.append(k_means.inertia_)
```

```python
import matplotlib.pyplot as plt

plt.plot(range(1,11),wcss_list)
plt.title("K-Means Algo")
plt.show()
```

```python
kmeans = KMeans(n_clusters = 5,init = 'k-means++',random_state  = 42 )
y_predict = kmeans.fit_predict(x)
```

```python
plt.scatter(x[y_predict == 0,0],x[y_predict == 0,1],s = 100,c = 'blue',label = 'cluster1')
plt.scatter(x[y_predict == 1,0],x[y_predict == 1,1],s = 100,c = 'red',label = 'cluster2')
plt.scatter(x[y_predict == 2,0],x[y_predict == 2,1],s = 100,c = 'green',label = 'cluster3')
plt.scatter(x[y_predict == 3,0],x[y_predict == 3,1],s = 100,c = 'yellow',label = 'cluster4')
plt.scatter(x[y_predict == 4,0],x[y_predict == 4,1],s = 100,c = 'blue',label = 'cluster5')
plt.show()
```
