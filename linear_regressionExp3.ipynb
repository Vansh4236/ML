```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
data = pd.read_csv("birthwt.csv")
X_age = data[['age']]
y_bwt = data['bwt']
```

```python
model1 = LinearRegression()
model1.fit(X_age, y_bwt)

age_range = np.linspace(X_age.min(), X_age.max(), 100)
bwt_pred = model1.predict(age_range)
```

```python
plt.figure()
plt.scatter(X_age, y_bwt)
plt.plot(age_range, bwt_pred)
plt.xlabel("Mother Age")
plt.ylabel("Birth Weight")
plt.title("Age vs Birth Weight")
plt.show()
```

```python
print("Age vs Birth Weight")
print("Coefficient:", model1.coef_[0])
print("Intercept:", model1.intercept_)
```

```python
y_lwt = data['lwt']

model2 = LinearRegression()
model2.fit(X_age, y_lwt)

lwt_pred = model2.predict(age_range)

plt.figure()
plt.scatter(X_age, y_lwt)
plt.plot(age_range, lwt_pred)
plt.xlabel("Mother Age")
plt.ylabel("Mother Weight")
plt.title("Age vs Mother Weight")
plt.show()

print("Age vs Mother Weight")
print("Coefficient:", model2.coef_[0])
print("Intercept:", model2.intercept_)
```
