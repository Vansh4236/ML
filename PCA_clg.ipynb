```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
iris = datasets.load_iris()
predictors = iris.data[:, 0:4]
target = iris.target
predictors = StandardScaler().fit_transform(predictors)
```

```python
pca = PCA(n_components=2)
```

```python
pc = pca.fit_transform(predictors)
```

```python
pc_ds = pd.DataFrame(data=pc,columns=['PC 1','PC 2'])
```

```python
pc_ds
```

```python
target_ds=pd.DataFrame(data=target,columns=['class'])
data_mod=pd.concat([pc_ds,target_ds],axis = 1)
```

```python
target_ds
```

```python
data_mod
```

```python
scatter = plt.scatter(data_mod['PC 1'], data_mod['PC 2'], c=data_mod['class'])
plt.xlabel("PC 1")
plt.ylabel("PC 2")
plt.title("PCA Plot")

plt.legend(*scatter.legend_elements(), title="Class")

plt.show()
```
